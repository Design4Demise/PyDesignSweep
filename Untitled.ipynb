```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('design_sweep_1609187517.csv.csv')
df.head()
```

```python
df2=pd.readcsv('design_sweep_1609191829.csv')
df2.head()
```

```python
df=df.dropna(how='any')
```

```python
df = df[df['BatteryMass'] != 'SPAR FAIL']
df = df[df['Aerofoil'] != 'RG15 (C)']
print(df.shape)
df.head()
```

```python
numeric_cols = ['AR','TaperRatio','SparHeight','SparWidth','BatteryMass','MaxStress']
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric)
```

```python
SF = 1.5
failure_stress = 4.4908e6 / SF 

df['WithinConstraints'] = df['MaxStress'] < failure_stress
df.head()
```

```python
failure_stress
```

```python
sorted_df=df.sort_values('BatteryMass', ascending=True)[df['WithinConstraints'] == 1]
```

```python
sorted_df.head(n=20)
```

```python
sorted_df['pcdiff'] = 100*(1- sorted_df['BatteryMass']/sorted_df['BatteryMass'][2089])
sorted_df.head(30)
```
